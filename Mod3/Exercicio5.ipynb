```python
import requests
url = 'https://api.covid19api.com/country/brazil'
rs = requests.get(url)

print(rs.status_code)
```

```python
dados = rs.json()
dados[0]
```

```python
dados_coletados = []
for registro in dados:
    conf = registro['Confirmed']
    if (int(conf) > 0): 
        dados_coletados.append([registro['Date'], registro['Confirmed']])
```

```python
DATA = 0
CONFIRMADOS = 1
for i in range(len(dados_coletados)):
    dados_coletados[i][DATA] = dados_coletados[i][DATA][:10]

dados_coletados
```

```python
dt_isolada = dados_coletados[0][0]
dt_isolada = dt_isolada.split("-")
print(dt_isolada)
```

```python
dia = dt_isolada[2]
mes = dt_isolada[1]
ano = dt_isolada[0]
```

```python
print(f'O primeiro caso de COVID no Brasil foi na data: {dia}/{mes}/{ano}')
```
